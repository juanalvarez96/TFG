```python
# Define function for extracting text out of the downloaded ironic and not ironic tweets
def get_text_noironicos(tweetid):
    tweetfile = 'DATA/noironicosjson/{}.json'.format(tweetid)
    print(tweetfile)
    if not os.path.exists(tweetfile):
        return 
    return json.load(open(tweetfile))['text']
```

```python
# Add body to database
df_ironicos['text']=df_ironicos['id_tweet'].apply(get_text)
df_noironicos['text']=df_noironicos['id_tweet'].apply(get_text_noironicos)
```
